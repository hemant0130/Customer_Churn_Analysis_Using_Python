# Customer Churn Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Customer_Churn.csv")
df
```

```python
df.info()
```

#replace blanks with 0 as tenure is 0 so no total charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def convert(value):
    if value == 1:
       return "yes"
    else:
       return "no"
        
df["SeniorCitizen"] = df["SeniorCitizen"].apply(convert)  
df.head()
```

#converted 0 and 1 values of senior citizens to yes/no to make it easier to understand

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(data = df,x = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by  Churn")
plt.show()
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
gb
plt.figure(figsize= (3,4))
plt.pie(gb["Churn"],labels = gb.index,autopct = "%.2f")
plt.title("Percentage of Churned Customers")
plt.show()
```

From the above pie chart we concluded that 26.54% customers are churned out.

```python
ax = sns.countplot(data = df,x="gender",hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Customers Churn by Gender")
plt.show()
```

From the above chart we concluded that number of churns in male category and female category having  slightly difference 

```python
plt.figure(figsize = (4,4))
ax=sns.countplot(data = df,x = "SeniorCitizen")
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
ax = sns.countplot(data = df,x="SeniorCitizen",hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title(" Churn by Senior Citizen")
plt.show()
```

From the above chart we conclude that churned senior citizens are less than non-churned senior citizens 

```python
plt.figure(figsize = (10,7))
sns.histplot(data = df,x= "tenure",bins = 72,hue = "Churn")
plt.title("Churn Based on Tenure")
plt.show()
```

#from above chart we concluded that people who have used our services for a long time have stayed and people who have used our 
services for 1 and 2 months have churned.

```python
plt.figure(figsize = (4,4))
ax=sns.countplot(data = df,x = "Contract",hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn of Customers by Contract")
plt.show()
```

 from the above chart we conclude that customers who have month-to-month contract are  more churned instead of customers who have 1 or 2 year contract

```python
plt.figure(figsize = (7,4))
ax=sns.countplot(data = df,x = "PaymentMethod",hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation = 45)
plt.show()
```

 From the above chart we concludes that customers who used electronic payment method are more churned in comparision to other payments methods
